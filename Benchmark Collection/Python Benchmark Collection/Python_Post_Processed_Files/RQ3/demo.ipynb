```python
import json

file_path = "/home/XXX/Tracing/Annotation/Python_Post_Processed_Files/RQ3/conditional_dataset_python.jsonl"

try:
    with open(file_path, 'r') as file:
        for line_number, line in enumerate(file, 1):
            try:
                entry = json.loads(line)
                print(f"\n--- Entry {line_number} ---")
                
                # Print Branch Code
                print("\nBranch Code:")
                print(entry.get('Source Code', 'No branch code available'))
                
                # Print Question
                print("\nQuestion:")
                print(entry.get('question', 'No question available'))
                
                # Print Answer
                print("\nAnswer:")
                print(entry.get('answer', 'No answer available'))
                print("-" * 40)
                
            except json.JSONDecodeError:
                print(f"\n⚠️ Error decoding JSON at line {line_number}")
                
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
import json

input_file = "/home/XXX/Traces/prompting/data/branch_prediction_dataset.jsonl"
output_file = "./conditional_dataset_python.jsonl"

with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for line in infile:
        try:
            entry = json.loads(line)
            
            # Remove unwanted key
            entry.pop("scratchpad_format", None)
            
            # Reorganize the entry with Source Code as second key
            modified_entry = {
                "idx": entry["idx"],
                "Source Code": entry.pop("Branch_Code"),
                **{k: v for k, v in entry.items() if k not in ["idx", "Branch_Code"]}
            }
            
            outfile.write(json.dumps(modified_entry) + '\n')
            
        except json.JSONDecodeError:
            print(f"Skipping invalid JSON line: {line.strip()}")
            continue
        except KeyError as e:
            print(f"Missing expected key {e} in entry: {entry.get('idx', 'unknown')}")
            continue

print(f"Reordered dataset saved to: {output_file}")
```
