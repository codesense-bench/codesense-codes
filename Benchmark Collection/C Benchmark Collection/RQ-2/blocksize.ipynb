```python
from collections import Counter
import json

# Initialize a counter
n_frequencies = Counter()

with open('c_block_analysis.jsonl', 'r') as file:
    for line in file:
        json_object = json.loads(line)
        n_value = json_object.get("Block_Size")
        n_frequencies[n_value] += 1

# Print frequency distribution
print("Block Size (n) Frequency Distribution:")
print("=" * 40)
for n, count in sorted(n_frequencies.items()):
    print(f"n = {n}: {count} samples ({count/sum(n_frequencies.values()):.1%})")

# Print total samples
print("\nTotal samples:", sum(n_frequencies.values()))
```

```python
import json
from tempfile import mkstemp
from shutil import move
from os import fdopen, remove

file_path = "c_block_analysis.jsonl"

# Create temp file
fh, abs_path = mkstemp()
with fdopen(fh, 'w') as new_file:
    with open(file_path) as old_file:
        for idx, line in enumerate(old_file):
            data = json.loads(line)
            data['idx'] = idx
            new_file.write(json.dumps(data) + '\n')

# Remove original file and replace with temp file
remove(file_path)
move(abs_path, file_path)
print(f"Added indices to {idx+1} records in-place")
```
