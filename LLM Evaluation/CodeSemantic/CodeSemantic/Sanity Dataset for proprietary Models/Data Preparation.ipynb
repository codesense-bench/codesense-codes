# Performance Analysis

```python
import json
from pathlib import Path
import re

input_path = Path("/home/XXX/CodeSemantic/CodeSemantic/Detailed_Results/Llama-3.1-8B-Instruct/output/1/Prompt_4_CoT_no_incontext_different_quantization_yes.jsonl")
response_path = Path("/home/XXX/CodeSemantic/CodeSemantic/Sanity Dataset for proprietary Models/Output/output_4_quantize_response.json")  

def clean_string(s):
    """Remove surrounding quotes and whitespace"""
    s = str(s).strip()
    # Remove surrounding single or double quotes if they exist
    if (s.startswith("'") and s.endswith("'")) or (s.startswith('"') and s.endswith('"')):
        s = s[1:-1]
    return s.strip()

def extract_answer(response):
    """Extract content from <ans> tags, or return the original string if no tags found"""
    match = re.search(r'<ans>(.*?)</ans>', str(response), re.IGNORECASE)
    return clean_string(match.group(1)) if match else clean_string(response)

def compare_answers(truth, response):
    """Robust comparison with debug output"""
    truth_clean = clean_string(truth)
    response_clean = clean_string(response)
    
    print(f"Comparison debug:")
    print(f"Original ground truth: {repr(truth)}")
    print(f"Original response: {repr(response)}")
    print(f"Cleaned ground truth: {repr(truth_clean)} (len:{len(truth_clean)})")
    print(f"Cleaned response: {repr(response_clean)} (len:{len(response_clean)})")
    print(f"Exact match: {truth_clean == response_clean}")
    return truth_clean == response_clean

# Read the response data
with open(response_path, 'r') as f:
    response_data = json.load(f)

correct_count = 0
total_samples = 0

with open(input_path, 'r') as f:
    for i, line in enumerate(f):
        if i >= 20:  
            break
            
        data = json.loads(line)
        idx = data.get('idx', 'N/A')
        ground_truth = data.get('ground_truth', 'N/A')
        
        response_idx = response_data[i]['idx']
        raw_response = response_data[i]['response'] if i < len(response_data) else 'N/A'
        response = extract_answer(raw_response)
        
        print(f"\nidx: {idx} and response idx:{idx}")
        is_correct = compare_answers(ground_truth, response)
        
        if is_correct:
            correct_count += 1
        total_samples += 1
        
        print(f"Ground truth: {repr(ground_truth)}")
        print(f"Raw response: {repr(raw_response)}")
        print(f"Extracted response: {repr(response)}")
        print(f"Correct: {'' if is_correct else ''}")
        print("-" * 60)

print(f"\nSummary:")
print(f"Total samples: {total_samples}")
print(f"Correct responses: {correct_count}")
print(f"Accuracy: {correct_count/total_samples:.2%}")
```
